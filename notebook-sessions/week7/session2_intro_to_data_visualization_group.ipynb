| **Labels** | Are all axes, titles, and legends present? | /5 |
| **Aesthetics** | Is it visually appealing? | /5 |
| **Fallback** | Does it degrade gracefully? | /5 |

### 🔑 Key Takeaways

| Chart Type | Best For | Avoid When |
|------------|----------|------------|
| **Bar** | Comparing categories | Continuous data |
| **Line** | Showing trends | Unrelated categories |
| **Scatter** | Relationships between variables | Few data points |
| **Pie** | Parts of a whole (≤5 slices) | Precise comparisons |
| **Histogram** | Distribution patterns | Categorical data |

### 💡 Visualization Golden Rules

1. **Start with the question**: What do you want to show?
2. **Choose the right chart**: Match the chart to your data
3. **Label everything**: Never assume the reader knows
4. **Keep it simple**: Avoid chartjunk and 3D effects
5. **Test your fallback**: Not everyone has matplotlib!
